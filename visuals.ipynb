```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Cleaned Data
try:
    df = pd.read_csv('C:/Users/vishn/OneDrive/Desktop/Cleaned_DataSet.csv')
    print("Data Loaded Successfully!")
except FileNotFoundError:
    print("Error: 'cleaned_data.csv' not found. Please check the file name.")

# --- GRAPH 1: HEATMAP (State vs Category) ---
# Instruction: "Explore relationships between two variables"
print("\nGenerating Heatmap (State vs Category)...")
plt.figure(figsize=(9, 6))

# Create a pivot table for the heatmap
pivot_table = df.pivot_table(index='state', columns='category', values='sales', aggfunc='sum')
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap: Which Category Sells Best in Which State?')
plt.show()

# --- GRAPH 2: SCATTER PLOT (Sales vs Profit) ---
if 'Profit' in df.columns:
    print("\nGenerating Scatter Plot (Sales vs Profit)...")
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x='sales', y='profit', data=df, hue='category', alpha=0.6)
    plt.title('Scatter Plot: Sales vs Profit (Correlation)')
    plt.show()
else:
    print("\n'Profit' column not found, skipping Scatter Plot.")

# --- GRAPH 3: BOX PLOT (Customer Segment) ---
print("\nGenerating Box Plot (Segment vs Sales)...")
plt.figure(figsize=(10, 6))
sns.boxplot(x='segment', y='sales', data=df, palette='Set2')
plt.ylim(0, 1000) 
plt.title('Box Plot: Sales Range by Customer Segment')
plt.show()
```
